```python

df = pd.read_csv("Local-nat-inter_070622_for BSC1_actualizada.csv", sep='\t')
df = df[['Local','National','International']]
df = df.fillna("NaN")
df = df.replace(to_replace= 'HCN NOS', value= 'NOS' )
df = df.replace(to_replace= 'HCN NOS || Other', value= 'NOS || Other' )
df = df.replace(to_replace= 'Rhabdo', value= 'Other' )
df = df.replace(to_replace= 'Ad', value= 'Other' )
df_counts = pd.DataFrame()

def counts(word, col):
    doubt = []
    for i in range(0, df.shape[0]):
        dff = df.loc[i][col]
        if word in df.loc[i][col]:
            doubt.append(dff)
            counter = pd.Series(doubt).value_counts()
    df_counts[col] = counter
            
            
    return(df_counts) 
    
counts("Other", "National")          
counts("Other", "Local")
counts("Other", "International")




df_counts.plot(kind="bar")
df_counts
```
